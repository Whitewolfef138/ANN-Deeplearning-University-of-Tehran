# Using MAE

```python
import tensorflow as tf
import keras 

class MLP_One(tf.keras.Model):

  def __init__(self):
    super(MLP_One, self).__init__()
    self.dense = keras.layers.Dense(2048 , activation='relu' , kernel_regularizer=keras.regularizers.L1L2(l1 = 0.001, l2 = 0.001))
    self.flatten=  keras.layers.Flatten()
    self.drop = keras.layers.Dropout(0.2)
    #self.bath_norm = keras.layers.BatchNormalization()
    self.output_layer = keras.layers.Dense(1 , activation='linear' )
    

  def call(self, inputs):
    x = self.dense(inputs)
    x = self.drop(x)
    #x = self.bath_norm(x)
    x = self.flatten(x)
    x = self.output_layer(x)
    return x
```

```python
input_layer = keras.layers.Input(scaled_df_train.iloc[0,:-1].shape)
output = MLP_One()(input_layer)
model_1 = keras.Model(inputs=input_layer, outputs=output)
model_1.compile(optimizer=keras.optimizers.SGD(momentum=0.1), loss=keras.losses.mean_absolute_error
                , metrics=[keras.metrics.R2Score , keras.metrics.mean_squared_error , keras.metrics.huber])
model_1.summary()
```

```python
history_1 = model_1.fit(x = scaled_df_train.iloc[:,:-1] , y = scaled_df_train.iloc[:,-1] , epochs=30, batch_size=64 , shuffle=True ,
            validation_data=(scaled_df_test.iloc[:,:-1],  scaled_df_test.iloc[:,-1]))
```

```python
plt.plot(history_1.history['loss'] , label = 'MAE train')
plt.plot(history_1.history['val_loss'] , label = 'MAE test')
plt.legend()
plt.tight_layout()
plt.title("One layer MAE")
plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.show()
plt.plot(history_1.history['mean_squared_error'] , label = 'MSE train')
plt.plot(history_1.history['val_mean_squared_error'] , label = 'MSE test')
plt.title("One layer MSE")
plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.plot(history_1.history['huber'] , label = 'huber train')
plt.plot(history_1.history['val_huber'] , label = 'huber test')
plt.title("One layer Huber")
plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.plot(history_1.history['r2_score'] , label = 'R2 train')
plt.plot(history_1.history['val_r2_score'] , label = 'R2 test')
plt.title("One layer R2-score")
plt.xlabel('Epoch')
plt.ylim([-1,+1])
plt.ylabel("R2-Score")
plt.legend()
plt.tight_layout()
plt.show()
```

```python


class MLP_two(tf.keras.Model):

  def __init__(self):
    super(MLP_two, self).__init__()
    self.dense_1 = keras.layers.Dense(256 , activation='relu' , kernel_regularizer=keras.regularizers.L1L2(l1 = 0.001, l2 = 0.001))
    self.dense_2 = keras.layers.Dense(512 , activation='relu' )
    self.flatten=  keras.layers.Flatten()
    self.drop = keras.layers.Dropout(0.15)
    #self.bath_norm = keras.layers.BatchNormalization()
    self.output_layer = keras.layers.Dense(1 , activation='linear')
    

  def call(self, inputs):
    x = self.dense_1(inputs)
    x = self.drop(x)
    x = self.dense_2(x)
    x = self.drop(x)
    x = self.flatten(x)
    x = self.output_layer(x)
    return x
```

```python
input_layer = keras.layers.Input(scaled_df_train.iloc[0,:-1].shape)
output = MLP_two()(input_layer)
model_2 = keras.Model(inputs=input_layer, outputs=output)
model_2.compile(optimizer=keras.optimizers.SGD(momentum=0.1), loss=keras.losses.mean_absolute_error
                , metrics=[keras.metrics.R2Score , keras.metrics.mean_squared_error , keras.metrics.huber])
model_2.summary()
```

```python
history_2 = model_2.fit(x = scaled_df_train.iloc[:,:-1] , y = scaled_df_train.iloc[:,-1] , epochs=100, batch_size=64 , 
            validation_data=(scaled_df_test.iloc[:,:-1],  scaled_df_test.iloc[:,-1]))
```

```python
plt.plot(history_2.history['loss'] , label = 'MAE train')
plt.plot(history_2.history['val_loss'] , label = 'MAE test')
plt.legend()
plt.tight_layout()
plt.title("Two layer MAE")
plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.show()
plt.plot(history_2.history['mean_squared_error'] , label = 'MSE train')
plt.plot(history_2.history['val_mean_squared_error'] , label = 'MSE test')
plt.title("Two layer MSE")
plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.plot(history_2.history['huber'] , label = 'huber train')
plt.plot(history_2.history['val_huber'] , label = 'huber test')
plt.title("Two layer Huber")
plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.plot(history_2.history['r2_score'] , label = 'R2 train')
plt.plot(history_2.history['val_r2_score'] , label = 'R2 test')
plt.title("Two layer R2-score")
plt.xlabel('Epoch')
plt.ylim([-1,+1])
plt.ylabel("R2-Score")
plt.legend()
plt.tight_layout()
plt.show()
```
